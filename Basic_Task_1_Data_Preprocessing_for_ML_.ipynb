```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
cancer_df = pd.read_csv('/content/Cancer_Data.csv')
```

```python
cancer_df.head()
```

```python
#checking if there is any null value present in the dataset
cancer_df.isnull().sum()
```

```python
cancer_df['diagnosis'].value_counts()
```

M = Malignant - Benign

B = Benign - Malignant

```python
#label encoding
encoder = LabelEncoder()
type_of_cancer = encoder.fit_transform(cancer_df.diagnosis)

#creating a new column named 'diagnosed_type' to distinguish the diagnosis based on label encoding
cancer_df['diagnosed_type'] = type_of_cancer
```

```python
cancer_df.head()
```

The encoding has been done according to alphabetical order which means the following:

Malignant --> 1

Benign --> 0

#Training and Test Split

```python
#dropped the 'diagnosis' column because i just created a label encoded column for the diagnosis so in training it will most probably just map M = 1 and B = 0, which is not properly training the model,
# also dropped id because it has no role in determining the type of cancer , it just represents a unique id of patient

x = cancer_df.drop(columns=['id', 'Unnamed: 32', 'diagnosis', 'diagnosed_type'], axis = 1)
y = cancer_df['diagnosed_type']
```
